```python
# 필요패키지 import
import numpy as np
import pandas as pd
import requests # 크롤링에 사용하는 패키지
from bs4 import BeautifulSoup # html 변환에 사용함
import time
import json # 동적 크롤링을 위한 json
import csv # csv 파일 저장

# 단일 종목 설정
code = '006400'

# 일일 주가데이터를 저장할 빈 리스트 생성
total_data = []

# 동적 크롤링을 위한 info 정의
info = {
    'referer': f'https://finance.daum.net/quotes/A{code}',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.61 Safari/537.36'
}

# page를 넘겨가면서 동적 크롤링
# 날짜 기준이 18.10.01~21.10.01(총 742건)
for page in range(1,75+1):
    
    seed = np.random.randint(100) # 시드도 난수로 만들고
    np.random.seed(seed) # 시드 생성
    a = np.random.randint(5) # 시드에서 난수 생성
    time.sleep(a)

    url = f'https://finance.daum.net/api/quote/A{code}/days?symbolCode=A{code}&page={page}&perPage=10&pagination=true'
    print(f'{code} 의 {page} 페이지 크롤링 중입니다.')

    response = requests.get(url, headers=info)

# json 데이터 파싱
    data = json.loads(response.text)
# 각 page의 데이터 total_data로 추가
    total_data.append(data['data'])

print(f'종목코드 : {code} 의 총 {page} 페이지를 크롤링 완료했습니다.')
```

```python
# 만들어진 total_data의 형태는 리스트 안의 리스트값으로 있어
# sum함수를 사용하여 내부 리스트 안의 data를 통합한다.
# concat으로는 코드 작성이 곤란하여 sum 실행
final_data = sum(total_data, [])
final_data
```
